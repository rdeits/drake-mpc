```python
import numpy as np
import matplotlib.pyplot as plt
```

## Early contact plot

```python
N = 10
planned_cs = np.ones(N)
planned_cs[0:3] = 0
adjusted_cs = np.ones(N)
plt.plot(planned_cs, color='b', linestyle='--', label='planned')
plt.plot(adjusted_cs, color='r', linestyle='--', label='adjusted')
plt.legend()
plt.xlabel('knot point')
plt.ylabel('contact indicator')
plt.show()
np.shape(planned_cs)
```

# MIQP potential contact assignments

```python
N = 10
planned_cs = np.ones(N)
idx_planned = 5
planned_cs[0:idx_planned] = 0
adjusted_cs = np.ones(N)

cm = plt.get_cmap('Reds')
color = cm(0.7)
plt.plot(planned_cs, color='b', linestyle='--', label='planned')

idx_min = 3;
idx_max = 8
idx_list = range(idx_min, idx_max)
def get_color(idx):
    color_min = 0.3
    color_max = 0.7
    v = color_min + (color_max - color_min)*(idx - idx_min)/(1.0*idx_max - idx_min)
    return cm(v)

for idx in idx_list:
    if idx == idx_planned:
        continue
        
    adjusted_cs = np.ones(N)
    adjusted_cs[0:idx] = 0
    color = get_color(idx)
    plt.plot(adjusted_cs, color=color, linestyle='--')
    
plt.legend()
plt.xlabel('knot point')
plt.ylabel('contact indicator')
plt.show()
np.shape(planned_cs)
```
